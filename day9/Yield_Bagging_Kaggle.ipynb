```python
from sklearn.ensemble import BaggingRegressor
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn import tree
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.metrics import roc_auc_score,accuracy_score,r2_score
from sklearn.model_selection import train_test_split, GridSearchCV,KFold
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer 
from sklearn.compose import make_column_selector
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv('train.csv', index_col=0)
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train.info()
```

```python
train.columns
```

```python
X= train.drop('yield', axis=1)
y=train['yield']
```

```python
lr = LinearRegression()
kfold = KFold(n_splits=5,shuffle=True,random_state=24)
bagg = BaggingRegressor(estimator=lr,random_state=24)
params= {'estimator' : [lr],'n_estimators':[10,20,50]}
gcv = GridSearchCV(bagg,param_grid=params,cv = kfold,scoring='r2')
```

```python
gcv.fit(X,y)
```

```python
bagg.get_params()
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
test= pd.read_csv("test.csv").set_index('id')
test.head(2)
```

```python
best_model= gcv.best_estimator_
best_model
```

```python
y_pred= best_model.predict(test)
```

```python
submit  = pd.read_csv('sample_submission.csv')
```

```python
submit
```

```python
submit['yield']= y_pred
```

```python
submit
```

```python
submit.to_csv('ss_9_nov_1_lr.csv',index=False)
```
