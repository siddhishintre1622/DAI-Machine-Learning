```python
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
import numpy as np
from sklearn.compose import make_column_transformer 
from sklearn.compose import make_column_selector
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score,log_loss
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

import warnings
warnings.filterwarnings("ignore") 
```

```python
satellite = pd.read_csv('Satellite.csv', sep= ";")
satellite.head(2)
```

```python
y= satellite['classes']
X= satellite.drop('classes', axis=1)
```

```python
#LDA
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=24,test_size=0.3,stratify=y)
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
#QDA
```

```python
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, y_train)
y_pred= qda.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

#LDA transform + K-NN

```python
knn = KNeighborsClassifier(n_neighbors=3)
lda = LinearDiscriminantAnalysis()
pipe= Pipeline([('LDA', lda), ('KNN',knn)])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
kfold = StratifiedKFold(n_splits=5, random_state=24,shuffle=True)
params = {'KNN__n_neighbors': np.arange(1,9)}
gcv = GridSearchCV(pipe, param_grid=params,cv=kfold)                   
```

```python
gcv.fit(X,y)
```

```python
print("Best Params:", gcv.best_params_)
print("Best Score:", gcv.best_score_)
```
