#Tuning for alpha values = [0.01,0.1,0.3,0.6,1,1.5,2,4,10]
Hyperparameter tuning
parameter optiimization

```python
alphas =  [0.01,0.1,0.3,0.6,1,1.5,2,4,10]
scores = []
for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X_train,y_train)
    y_pred = ridge.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
    
```

```python
scores
```

```python
np.max(scores)
```

```python
i_max = np.argmax(np.max(scores))
```

```python
print("Best alpha" , alphas[i_max])
```

```python
print("Best Scores :",scores[i_max])
```

```python
alphas =  np.linspace(0.0001,10,20)
scores = []
for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X_train,y_train)
    y_pred = ridge.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
    
```

```python
np.max(scores)
```
