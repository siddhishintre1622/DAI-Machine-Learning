Constant Stragey

```python
imp_const = SimpleImputer(strategy='constant', fill_value=0.5).set_output(transform='pandas')
chem_const = imp_const.fit_transform(chem)
chem_const.isnull().sum().sum()
```

Median Strategy

```python
imp_med = SimpleImputer(strategy='median', fill_value=0.5).set_output(transform='pandas')
chem_const = imp_med.fit_transform(chem)
chem_const.isnull().sum().sum()
```

Mean Strategy

```python
imp_mean = SimpleImputer(strategy='mean', fill_value=0.5).set_output(transform='pandas')
chem_const = imp_mean.fit_transform(chem)
chem_const.isnull().sum().sum()
```

```python
X = chem.drop('Yield',axis=1)
y = chem['Yield']
```

```python
X_train, X_test, y_train, X_tst, = train_test_split(X,y, random_state=24, test_size=0.3)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_trn_imp,y_train)
```

```python
y_pred = lr.predict(X_tst_imp)
```
